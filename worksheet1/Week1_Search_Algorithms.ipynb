## 1. Breadth-First Search (BFS)

```python
from collections import deque

def bfs(graph, start, goal):
    """
    Breadth-First Search — explores all nodes at current depth before going deeper.
    Guarantees shortest path in unweighted graphs.
    Time: O(V + E), Space: O(V)
    """
    queue = deque([[start]])  # queue of paths
    visited = set([start])
    
    while queue:
        path = queue.popleft()
        node = path[-1]
        
        if node == goal:
            return path
        
        for neighbour in graph.get(node, []):
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(path + [neighbour])
    
    return None  # No path found

# Test graph
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E', 'G'],
    'G': ['F']
}

path = bfs(graph, 'A', 'G')
print(f'BFS path from A to G: {path}')
print(f'Path length: {len(path) - 1} steps')
```

## 2. Depth-First Search (DFS)

```python
def dfs(graph, start, goal, path=None, visited=None):
    """
    Depth-First Search — explores as deep as possible before backtracking.
    Does NOT guarantee shortest path.
    Time: O(V + E), Space: O(V)
    """
    if path is None: path = [start]
    if visited is None: visited = set([start])
    
    if start == goal:
        return path
    
    for neighbour in graph.get(start, []):
        if neighbour not in visited:
            visited.add(neighbour)
            result = dfs(graph, neighbour, goal, path + [neighbour], visited)
            if result:
                return result
    
    return None

path_dfs = dfs(graph, 'A', 'G')
print(f'DFS path from A to G: {path_dfs}')
print(f'Path length: {len(path_dfs) - 1} steps')
```

## 3. A* Search Algorithm

```python
import heapq

def astar(graph_weighted, heuristic, start, goal):
    """
    A* Search — uses f(n) = g(n) + h(n)
    g(n) = actual cost from start to n
    h(n) = heuristic estimate from n to goal
    Guarantees optimal path if heuristic is admissible.
    """
    # Priority queue: (f_cost, g_cost, node, path)
    open_set = [(heuristic[start], 0, start, [start])]
    visited = {}
    
    while open_set:
        f, g, node, path = heapq.heappop(open_set)
        
        if node == goal:
            return path, g
        
        if node in visited and visited[node] <= g:
            continue
        visited[node] = g
        
        for neighbour, cost in graph_weighted.get(node, []):
            new_g = g + cost
            new_f = new_g + heuristic.get(neighbour, 0)
            heapq.heappush(open_set, (new_f, new_g, neighbour, path + [neighbour]))
    
    return None, float('inf')

# Weighted graph
weighted_graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3)],
    'D': [('G', 6)],
    'E': [('G', 2)],
    'F': [('G', 4)],
    'G': []
}

# Heuristic (estimated distance to goal G)
h = {'A': 7, 'B': 6, 'C': 5, 'D': 4, 'E': 2, 'F': 4, 'G': 0}

path_astar, cost = astar(weighted_graph, h, 'A', 'G')
print(f'A* path from A to G: {path_astar}')
print(f'Total cost: {cost}')
```

## 4. Comparison

```python
import matplotlib.pyplot as plt

algorithms = ['BFS', 'DFS', 'A*']
path_lengths = [len(bfs(graph, 'A', 'G')) - 1,
                len(dfs(graph, 'A', 'G')) - 1,
                len(path_astar) - 1]

plt.figure(figsize=(8, 4))
bars = plt.bar(algorithms, path_lengths, color=['#0EA5E9', '#6366F1', '#10B981'])
plt.title('Search Algorithm Path Lengths Comparison')
plt.ylabel('Path Length (steps)')
for bar, val in zip(bars, path_lengths):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05, str(val), ha='center')
plt.tight_layout()
plt.show()

print('\nAlgorithm Comparison:')
print(f'BFS: Shortest path guaranteed, explores level by level')
print(f'DFS: May find longer path, uses less memory')
print(f'A*:  Optimal with admissible heuristic, most efficient')
```
