## 1. NumPy Basics

```python
import numpy as np
import matplotlib.pyplot as plt

# Create arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.arange(0, 10, 2)
matrix = np.zeros((3, 3))

print('Array 1:', arr1)
print('Array 2:', arr2)
print('3x3 Zero Matrix:\n', matrix)
```

```python
# Array operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print('Sum:', a + b)
print('Dot product:', np.dot(a, b))
print('Mean of a:', np.mean(a))
print('Std of b:', np.std(b))
```

## 2. Pandas for Data Handling

```python
import pandas as pd

# Create a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 22],
    'Score': [85, 92, 78, 95, 88]
}

df = pd.DataFrame(data)
print(df)
print('\nBasic Statistics:')
print(df.describe())
```

```python
# Data filtering and manipulation
high_scorers = df[df['Score'] > 85]
print('High scorers (>85):')
print(high_scorers)

# Sorting
sorted_df = df.sort_values('Score', ascending=False)
print('\nSorted by score:')
print(sorted_df)
```

## 3. Data Visualization

```python
import matplotlib.pyplot as plt

# Bar chart
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(df['Name'], df['Score'], color=['#0EA5E9', '#6366F1', '#10B981', '#F59E0B', '#EF4444'])
axes[0].set_title('Student Scores')
axes[0].set_xlabel('Student')
axes[0].set_ylabel('Score')

# Scatter plot
axes[1].scatter(df['Age'], df['Score'], color='#6366F1', s=100)
axes[1].set_title('Age vs Score')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Score')

plt.tight_layout()
plt.show()
```

## 4. Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Sample data with different scales
data = np.array([[100, 0.5], [200, 1.5], [300, 2.5], [400, 3.5]])

# Standard scaling (mean=0, std=1)
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
print('Standard Scaled:')
print(scaled)

# Min-Max scaling (0 to 1)
mm_scaler = MinMaxScaler()
mm_scaled = mm_scaler.fit_transform(data)
print('\nMin-Max Scaled:')
print(mm_scaled)
```
